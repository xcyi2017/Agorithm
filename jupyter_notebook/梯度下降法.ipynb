```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'SimSun'
matplotlib.rcParams['font.size'] = 10

x = np.linspace(-10, 10, 100)
f = lambda x: x ** 2 - 3 * x -10
prime_f = lambda x: 2 * x - 3
y = f(x)

plt.figure()
plt.plot(x, y)
plt.title('测试')
plt.xlabel('这是$x^2$')
plt.ylabel('这是$y^2$')
plt.show()
```

```python
# 梯度下降法
def gradient(data, max_iter=100, alpha=0.2, err=1e-4):
    """
    data: 传入数据
    alpha: 学习速率
    max_iter: 最大迭代次数
    err: 允许误差
    """
    r = 0
    trace_x = []
    current_x_data = data
    while r < max_iter:
        current_y_data = f(current_x_data)
        trace_x.append(current_x_data)
        current_x_data = current_x_data - alpha / (r+1) * prime_f(current_x_data)
        if np.abs(current_y_data - prime_f(current_x_data)) <= err:
            break
        r += 1
    trace_x.append(current_x_data)
    return np.array(trace_x)

fig = plt.figure(figsize=(5,4))
ax = fig.add_subplot(111)
ax.plot(x, y)
x_ = gradient(np.random.uniform(-10,10))
ax.scatter(x_, f(x_), color='g', marker='*')
ax.set_xlabel('x轴')
ax.set_ylabel('y轴')
plt.show()
```
